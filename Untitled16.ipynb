```python
'''Design and develop programs to implement the Depth-First Search (DFS) 
algorithm for graph traversal in Artificial Intelligence with the following 
specifications: 
i. Construct a static graph consisting of 6 nodes and 10 edges, 
where the graph is stored using a dictionary representation. 
The graph is given as: A- B, A-C, B-D, B-E, C-D, C-F, D-E, 
D-F, E-F, B-F.  
ii. Create another graph where the number of nodes and number 
of edges are provided by the user at runtime. 
iii. Accept the initial node and the goal node from the user. 
iv. Display the path from the initial node to the goal node for both 
the cases. '''
```

```python
def dfs(graph, start, goal, visited=None):
    if visited is None:
        visited = set()

    print(start, end=" ")

    if start == goal:
        return True

    visited.add(start)

    for neighbor in graph.get(start, []):
        if neighbor not in visited:
            if dfs(graph, neighbor, goal, visited):
                return True

    return False

static_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

print("=== Static Graph DFS ===")
start_node = input("Enter the initial node: ").upper()
goal_node = input("Enter the goal node: ").upper()
print("DFS traversal:")
dfs(static_graph, start_node, goal_node)

# Graph input
graph = {}
n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbours separated by space: ").split()
    graph[node] = neighbors


start = input("Enter start node: ")
goal = input("Enter goal node: ")

print("DFS traversal:")
dfs(graph, start, goal)
```

```python
'''Design and develop programs to implement the Breadth-First Search (BFS) 2 
2025-26 onwards 
algorithm for graph traversal in Artificial Intelligence with the following 
specifications: 
i. Construct a static graph consisting of 6 nodes and 10 edges, where the 
graph is stored using a dictionary representation. 
The graph is given as: 
A–B, A–C, B–D, B–E, C–D, C–F, D–E, D–F, E–F, B–F. 
ii. Create another graph dynamically, where the number of nodes and 
number of edges are provided by the user at runtime, and the user specifies 
which nodes are connected. 
iii. Accept the initial node and the goal node from the user. 
iv. Perform Breadth-First Search (BFS) traversal starting from the given 
initial node and: 
v. Display the order of node traversal 
vi. Display the shortest path from the initial node to the goal node 
for both the static and dynamic graph cases.'''
```

```python
# -------------------- BFS FUNCTION --------------------
def bfs(graph, start, goal):
    visited = set()
    queue = [[start]]  

    traversal = []

    while queue:
        path = queue.pop(0)   
        node = path[-1]

        if node not in visited:
            visited.add(node)
            traversal.append(node)

            if node == goal:
                return traversal, path

            for neighbour in graph[node]:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

    return traversal, None
    
static_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

print("----- STATIC GRAPH -----")
start = input("Enter start node: ").upper()
goal = input("Enter goal node: ").upper()

traversal, path = bfs(static_graph, start, goal)

print("Traversal Order:", traversal)
print("Shortest Path:", path)

print("\n----- DYNAMIC GRAPH -----")

n = int(input("Enter number of nodes: "))
nodes = []

for i in range(n):
    node = input(f"Enter node {i+1}: ").upper()
    nodes.append(node)

graph = {}
for node in nodes:
    graph[node] = []

e = int(input("Enter number of edges: "))

for i in range(e):
    u = input("Enter first node: ").upper()
    v = input("Enter second node: ").upper()

    graph[u].append(v)
    graph[v].append(u)   

start = input("Enter start node: ").upper()
goal = input("Enter goal node: ").upper()

traversal, path = bfs(graph, start, goal)

print("Traversal Order:", traversal)
print("Shortest Path:", path)
```

```python
'''Construct a program to implement the A* search algorithm using heuristic 
values for a given graph. The graph is stored using a dictionary 
representation. The graph consists of the following nodes and edge costs: 
Path Edge Cost 
A-B 1 
A-C 3 
B-D 3 
B-E 6 
C-D 1 
D-E 1 
 
Node A B C D E 
h(n) 7 6 4 2 0 
 
Accept the initial node from the user and node E as goal node. Apply the 
A* algorithm to: 
i. Display the order of node expansion 
ii. Display the optimal path 
iii. Display the total path cost
```

```python
import networkx as nwx

graph = nwx.Graph()

while True:
    choice = int(input("Enter choice (1 to add edge, 0 to stop): "))
    if choice == 1:
        node1 = input("Enter node 1: ").strip().upper()
        node2 = input("Enter node 2: ").strip().upper()
        weight = int(input("Enter weight: "))
        graph.add_edge(node1, node2, weight=weight)
    else:
        break

print(graph.edges(data=True))

start = input("Enter start node: ").strip().upper()
goal = input("Enter end node: ").strip().upper()


heuristic_values = {}
heuristic_values[goal] = 0

for node in graph.nodes():
    if node != goal:
        heuristic = int(input(f"Enter heuristic for {node}: "))
        heuristic_values[node] = heuristic

def get_heuristic(u, v):
    return heuristic_values[u]

path = nwx.astar_path(graph, start, goal, heuristic=get_heuristic, weight='weight')

print("Optimal Path:", "--->".join(path))

total_cost = nwx.astar_path_length(graph, start, goal, heuristic=get_heuristic, weight='weight')
print("Total Cost:", total_cost)
print("Node Expansion Order:", path)
```

```python
'''Consider a dataset containing records related to customer purchasing 
behavior. The dataset includes multiple numerical features describing 
customer demographics and spending patterns.The program should perform 
the following tasks: 
 
i. Read the given dataset into a dataframe. 
ii. Perform necessary data preprocessing and feature scaling. 
iii. Apply the K-Means Clustering algorithm on the dataset for different 
values of K. 
iv. Determine the optimal number of clusters using the Elbow Method. 
v. Validate the clustering results using the Silhouette Score. 
vi. Perform clustering using the optimal value of K and assign cluster labels 
to the data points. 
vii. Visualize the formed clusters and analyze the clustering results. '''
```

```python
import pandas as pd
data = pd.read_csv("customer_data.csv")

print("Dataset Loaded")
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
print("Data Scaling Completed")
```

```python
from sklearn.cluster import KMeans

k_values = [2, 3, 4, 5]

print("K-Means Clustering Results:\n")

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled_data)
    
    print(f"K = {k}")
    print("Inertia (WCSS):", kmeans.inertia_)
    print("-" * 30)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Inertia)")
plt.grid()
plt.show()

print("Conclusion:")
print("The optimal K is where the elbow bends (sharp decrease stops).")
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

optimal_k = 3  

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
labels = kmeans.fit_predict(scaled_data)

score = silhouette_score(scaled_data, labels)

print("Silhouette Score:", score)

print("\nConclusion:")
print("Higher silhouette score indicates better and well-separated clusters.")
```

```python
data["Cluster"] = labels

print("Cluster Assignment Completed")
print(data.head())

print("\nConclusion:")
print("Each data point is assigned to a cluster based on similarity.")
```

```python
plt.scatter(
    data["AnnualIncome"],
    data["SpendingScore"],
    c=data["Cluster"],
    cmap="viridis"
)

plt.title("Customer Clusters Visualization")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.grid()
plt.show()

print("\nConclusion:")
print("Clusters clearly show different customer groups based on spending behavior and income levels.")
```

```python
''You are provided with a transactional market basket dataset 
(market_basket_dataset.csv), where each transaction contains a set of items 
purchased together by a customer. Develop a program to demonstrate 
Association Rule Mining using the Apriori Algorithm. 
The program should perform the following tasks: 
 
i. Read the transactional dataset from a CSV file. 
ii. Convert the dataset into a suitable format for association rule mining. 
iii. Apply the Apriori algorithm to generate frequent itemsets based on a 
minimum support threshold. 
iv. Generate association rules from the frequent itemsets using minimum 
confidence. 
v. Display the discovered frequent itemsets and association rules along with 
their support, confidence values. 
```

```python
import pandas as pd

data = pd.read_csv("market_basket_dataset.csv")

print("Dataset Loaded Successfully")
print(data.head())
```

```python
transactions = []

for i in range(len(data)):
    items = str(data["Items"][i]).split(",")
    transactions.append(items)

print("Transactions:")
print(transactions)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print("One-Hot Encoded DataFrame:")
print(df.head())

frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)

print("\nFrequent Itemsets:")
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print("Association Rules Generated Successfully")
```

```python
print("\n--- FREQUENT ITEMSETS ---")
print(frequent_itemsets)

print("\n--- ASSOCIATION RULES ---")
print(rules[["antecedents", "consequents", "support", "confidence"]])

print("\n--- FINAL READABLE RULES ---")

for i in range(len(rules)):
    print(
        list(rules.iloc[i]["antecedents"]),
        "->",
        list(rules.iloc[i]["consequents"]),
        "| support:", round(rules.iloc[i]["support"], 2),
        "| confidence:", round(rules.iloc[i]["confidence"], 2)
    )

print("\nConclusion:")
print("Apriori algorithm finds frequent itemsets and generates association rules based on support and confidence values.")
```
